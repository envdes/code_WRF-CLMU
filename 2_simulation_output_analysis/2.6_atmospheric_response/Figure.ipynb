# Plot temperature at atmospheric levels
- This script is used to plot temperature at atmospheric levels
- Simulations: GM_SLUCM, GM_CLMU, GM_CLMU_ALB, GM_CLMU_NOAH. 

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
```

```python
df_time_mean = pd.read_pickle('./data_for_figure/vertical_profile_mean.pkl')
df_time_mean
```

```python
len(df_time_mean.loc[0]['T_actual'])
```

```python
season_list = ['summer', 'winter']
nrows = 1
ncols = 4
lc1 = '#1f78b4'
lc2 = '#e31a1c'
lc3 = '#1b9e77'
lc4 = '#7570b3'
lc_0 = [lc1, lc2, lc2, lc3, lc4]
alpha_list_0 = [1, 1, 0.7, 1, 1]
labelcolor = '#6b6b6b'
fontsize = 6
title_list = ['Summer 13:30', 'Summer 01:30', 'Winter 13:30', 'Winter 01:30']
boundarywidth = 0.5
labelfont = 6
pad = 2
xwidth = 0.5
length = 2
linewidth = 0.5
padding = 4
handletextpad = 0.005
ax_xmin_list = [20, 16, -4, -4]
ax_xmax_list = [32, 22, 4, 4]
ax2_xmin_list = [-1, -2, -1, -1]
lc = [lc1, lc2, lc3, lc4]
alpha_list = [1, 1, 0.7, 1]
model_list = ['wrf', 'wrf-ctsm', 'wrf-ctsm_road_albedo', 'wrf-ctsm_uniform_urban']
label_list = ['GM_SLUCM', 'GM_CLMU', 'GM_CLMU$_\mathrm{ALB}$', 'GM_CLMU$_\mathrm{NOAH}$'] #
```

```python
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5.5, 3.5))
ax = axes.flatten()

for i in range(nrows * ncols):
    idx_time = i % 2
    tag = ['13:30', '01:30'][idx_time]
    df_time_tag = df_time_mean[df_time_mean['tag'] ==tag]
    idx_location = i // 2
    season = season_list[idx_location]
    df_time_tag_location = df_time_tag[df_time_tag['season'] == season]
    print(f'Plotting {season} {tag} ...')
    ax2 = ax[i].twiny()
    for m, model in enumerate(model_list):
        df_time_model = df_time_tag_location[df_time_tag_location['model'] == model]
        if m ==0:
            ax[i].scatter(df_time_model['T_actual'].values[0], df_time_model['z'].values[0], color = lc[m], alpha=alpha_list[m], label=label_list[m], s=5,
                      edgecolor='none')
            ax[i].plot(df_time_model['T_actual'].values[0], df_time_model['z'].values[0], color=lc[m], alpha=0.4, linewidth=linewidth)
            for l in range(len(df_time_model['T_actual'].values[0])):
                ax[i].axhline(y=df_time_model['z'].values[0][l], color=labelcolor, linestyle='--', linewidth=linewidth, alpha=0.4)
        else:
            diff = df_time_model['T_actual'].values[0] - df_time_tag_location[df_time_tag_location['model'] == 'wrf']['T_actual'].values[0]
            ax2.scatter(diff, df_time_model['z'].values[0], color = lc[m], alpha=alpha_list[m], label=label_list[m]+' minus GM_SLUCM', s=5,
                        edgecolor='none')
            ax2.plot(diff, df_time_model['z'].values[0], color=lc[m], alpha=0.2, linewidth=linewidth, linestyle='--')
            ax2.fill_betweenx(df_time_model['z'].values[0], 0, diff, color=lc[m], alpha=0.1, edgecolor='none')
    ax2_min = ax2_xmin_list[i]
    ax2_max = -ax2_xmin_list[i]
    ax2.set_xlim(ax2_min, ax2_max)
    xticks2 = [ax2_xmin_list[i], 0, ax2_max]
    ax2.set_xticks(xticks2)
    ax2.set_xticklabels([f'{tick:.0f}' for tick in xticks2])
    ax[i].set_xlim(ax_xmin_list[i], ax_xmax_list[i])
    xticks = [ax_xmin_list[i], (ax_xmin_list[i]+ax_xmax_list[i])/2, ax_xmax_list[i]]
    ax[i].set_xticks(xticks)
    ax[i].set_xticklabels([f'{tick:.0f}' for tick in xticks])
    if i ==0:
        lines2, labels2 = ax2.get_legend_handles_labels()
    ax[i].set_xlabel('TA (C)', fontsize=fontsize, color=lc1, labelpad=2)
    ax2.set_xlabel(r'$\Delta$TA (C)', fontsize=fontsize, color=labelcolor, labelpad=2)
    #ax[i].set_title(f'({string.ascii_lowercase[i]}) {title_list[i]}', fontsize=fontsize, color=labelcolor, pad=-padding)
    ax[i].text(0.05, 0.965, f'({string.ascii_lowercase[i]}) {title_list[i]}',transform=ax[i].transAxes,color=labelcolor,fontsize=fontsize)
    #ax[i].set_yticks(np.arange(50, 1050, 100))
    for spine in ax[i].spines.values():
        spine.set_linewidth(boundarywidth)
        spine.set_edgecolor(labelcolor)
    ax[i].spines['bottom'].set_color(lc1)   
    ax[i].spines['top'].set_color(labelcolor)  
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.spines['bottom'].set_visible(False)
    ax2.spines['left'].set_visible(False)
    if i ==0:
        labelleft= True    
        ax[i].set_ylabel(r'$z$ (m)', fontsize=fontsize, color=labelcolor, labelpad=0) # effective height above the displacement plane
    else:
        labelleft = False    
    ax[i].tick_params(axis='y', which='both', left=True, right=False, labelleft=labelleft, labelcolor=labelcolor, color=labelcolor,
                      labelsize=labelfont,pad=pad, width=xwidth, length=length)
    ax[i].tick_params(axis='x', which='both', top=False, bottom=True, labelbottom=True, labelcolor=lc1,color=lc1,
                      labelsize=labelfont,pad=pad, width=xwidth, length=length)
    ax2.tick_params(axis='x', labelbottom=False, labeltop=True, labelsize=labelfont, length=length, width=xwidth, labelcolor=labelcolor, colors=labelcolor, pad=pad/2)   
    ax[i].grid(axis='x', which='both', linestyle='--', linewidth=linewidth, alpha=0.2)
    ax[i].set_ylim(50, 1000)  
   
lines1, labels1 = ax[0].get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
legend = fig.legend(lines, labels, bbox_to_anchor=(0.5, -0.015), loc='lower center', frameon=False, 
                        handletextpad=handletextpad/10, fontsize=labelfont-1, ncol=4)
for i, text in enumerate(legend.get_texts()):
    text.set_color(lc[i])
    text.set_alpha(alpha_list[i])

plt.subplots_adjust(left=0.065, right=0.99, bottom=0.1, top=0.93, wspace=0.1, hspace=0.25) 
plt.savefig('vertical_temperature_profile.png', dpi=300)  
plt.savefig('vertical_temperature_profile.pdf', dpi=600) 
```
