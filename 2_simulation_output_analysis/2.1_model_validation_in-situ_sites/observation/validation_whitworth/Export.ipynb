# Export Whitworth site data

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
home_path = "/gws/nopw/j04/duicv/yuansun/"
```

```python
start_time = pd.Timestamp('2022-01-01 00:00:00')
end_time = pd.Timestamp('2022-12-31 23:00:00')
station_name = 'whitworth'
```

# Observation

```python
df_obs = pd.read_csv(f'{home_path}manchester/whitworth/hourly_met/hourly_data/existing_hourly.csv')
df_obs['Datetime'] = pd.to_datetime(df_obs['Datetime'])
df_obs.rename(columns={'Datetime': 'time'}, inplace=True)
df_obs
```

```python
df_obs_sel = df_obs[(df_obs['time'] >= start_time) & (df_obs['time'] <= end_time)]
df_obs_sel['time'] = pd.to_datetime(df_obs_sel['time'])
df_obs_sel = df_obs_sel.sort_values('time').reset_index(drop=True)
df_obs_sel_sel = df_obs_sel[['time', 'TBOT', 'RH', 'WIND', 'RAIN']]
df_obs_sel_sel.rename(columns={'TBOT': 'TAS'}, inplace=True)
df_obs_sel_sel.to_csv(f'./data_for_figure/{station_name}_obs.csv', index=False)
df_obs_sel_sel
```
