```python
sensor_id = 'MOD-PM-00427'
meta_sensor = meta[meta['sn'] == sensor_id]
meta_sensor
```

```python
df_sensor = pd.read_csv(f'{home_path}0_lcz_mcr/output_analysis/single_point/calibration/adjusted_sensor_data/{sensor_id}.csv')
df_sensor 
```

```python
df_sensor_clean = df_sensor[['timestamp', 'rh_hourly_avg_corrected', 'temp_hourly_avg_corrected']].copy()
df_obs = df_sensor_clean.rename(columns={'timestamp': 'time', 
                                         'rh_hourly_avg_corrected': 'RH',
                                         'temp_hourly_avg_corrected': 'TAS'})
df_obs['time'] = pd.to_datetime(df_obs['time'])
df_obs_sel = df_obs[(df_obs['time'] >= start_time) & (df_obs['time'] <= end_time)]
df_obs_sel.reset_index(drop=True, inplace=True)
df_obs_sel.to_csv(f'./data_for_figure/{station_name}_obs.csv', index=False)
df_obs_sel
```
