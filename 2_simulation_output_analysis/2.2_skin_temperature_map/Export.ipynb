# Export skin temperature on 16 July 2022
- This script is used to export skin temperature on 16 July 2022 for Greater Manchester.
- Simulations: GM_SLUCM, GM_CLMU. 
- Reference data: MODIS, VIIRS.

```python
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
home_path = '/gws/nopw/j04/duicv/yuansun/'
```

```python
ds_wrf_ctsm = xr.open_dataset(f'{home_path}0_WRFvsWRF-CTSM/archive/d04_wrf-ctsm/wrfout_d01_2022-07-15_01:00:00.nc')
#for date in ['07-16', '07-17', '07-18', '07-19']:
for date in ['07-16' , '07-19']:
    day_time = pd.to_datetime(f'2022-{date} 13:00:00')
    night_time = pd.to_datetime(f'2022-{date} 02:00:00')
    time = pd.to_datetime(ds_wrf_ctsm['XTIME'].values)
    idx_day = np.where(time == day_time)[0].item()
    idx_night = np.where(time == night_time)[0].item()
    wrf_cstm_day = (ds_wrf_ctsm['TSK']-273.15)[idx_day:idx_day+2].mean(dim='Time')
    wrf_cstm_night = (ds_wrf_ctsm['TSK']-273.15)[idx_night:idx_night+2].mean(dim='Time')
    wrf_cstm_day.to_netcdf(f'./data_for_figure/wrf-ctsm_day_{date}.nc')
    wrf_cstm_night.to_netcdf(f'./data_for_figure/wrf-ctsm_night_{date}.nc')
```

```python
for date in ['07-16', '07-19']:
    day_time = pd.to_datetime(f'2022-{date} 13:00:00')
    night_time = pd.to_datetime(f'2022-{date} 02:00:00')
    if date =='07-16':
        ds_wrf = xr.open_dataset(f'{home_path}0_WRFvsWRF-CTSM/archive/d04_wrf/wrfout_d01_2022-07-13_01:00:00.nc')
    else:
        ds_wrf = xr.open_dataset(f'{home_path}0_WRFvsWRF-CTSM/archive/d04_wrf/wrfout_d01_2022-07-18_01:00:00.nc')   
    time = pd.to_datetime(ds_wrf['XTIME'].values)
    idx_day = np.where(time == day_time)[0].item()
    idx_night = np.where(time == night_time)[0].item()
    wrf_day = (ds_wrf['TSK']-273.15)[idx_day:idx_day+2].mean(dim='Time')
    wrf_night = (ds_wrf['TSK']-273.15)[idx_night:idx_night+2].mean(dim='Time')
    wrf_day.to_netcdf(f'./data_for_figure/wrf_day_{date}.nc')
    wrf_night.to_netcdf(f'./data_for_figure/wrf_night_{date}.nc')
```
